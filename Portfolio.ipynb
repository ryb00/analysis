```python
gender = df.groupby(by = 'Gender')['Gender'].count()
gender.plot.pie(y='Gender', autopct='%1.1f%%')
```

```python
age = df.groupby(by = 'Age')['Age'].count()
age.plot.area(ylabel = 'Number')
age
```

```python
age_gender = df.groupby(by = ['Age','Gender'])['Age'].count()
#age_gender.plot.line(subplots = True, ylabel = 'Number')
age_female = df[df['Gender'] == 'Female'][['Age','Gender']].groupby(by = 'Age').count()
age_male = df[df['Gender'] == 'Male'][['Age','Gender']].groupby(by = 'Age').count()
age_gender_total = age_female.merge(age_male, how = 'inner', on = 'Age')
age_gender_total = age_gender_total.rename(columns={'Gender_x': 'Female', 'Gender_y': 'Male'})
age_gender_total.plot.line(ylabel = 'Number', title = 'Number of female and male passengers')
```

```python
age_gender_total
```

```python
age_gender_total['difference'] = age_gender_total['Female'] - age_gender_total['Male'] 
```

```python
age_gender_total
```

```python
age_gender_total['difference'].plot.line(ylabel = 'Number', title = 'Difference in the number of female and male passengers')
```

```python
flight_status_cancelled = df[df['Flight Status'] == 'Cancelled']
flight_status_cancelled_airport = flight_status_cancelled.groupby(by = 'Airport Name')['Airport Name'].count().sort_values(ascending=False).head(20)
flight_status_cancelled_airport.plot.barh(title = 'Cancelled')
```
